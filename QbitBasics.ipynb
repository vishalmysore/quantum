```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1)  # 1 qubit, no classical bits yet

print(qc.draw())
```

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1, 1)

qc.x(0)           # Flip |0вҹ© вҶ’ |1вҹ©
qc.measure(0, 0)  # Measure qubit 0, store in classical bit 0

print(qc.draw())
```
