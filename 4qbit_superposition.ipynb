```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer

# Create a quantum circuit with 4 qubits and 4 classical bits
qc = QuantumCircuit(4, 4)

# Step 1: Put all 4 qubits into superposition
qc.h(0)
qc.h(1)
qc.h(2)
qc.h(3)

# Step 2: Apply X gate to qubit 2
qc.x(2)

# Step 3: Measurement
qc.measure([0,1,2,3], [0,1,2,3])

# Run simulation
backend = Aer.get_backend('qasm_simulator')

compiled = transpile(qc, backend)
job = backend.run(compiled, shots=1024)
result = job.result()

counts = result.get_counts()
print(counts)
```

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer

# 4 qubits, 4 classical bits
qc = QuantumCircuit(4, 4)

# Step 1: Put first qubit in superposition
qc.h(0)

# Step 2: Entangle all qubits with qubit 0
qc.cx(0, 1)
qc.cx(1, 2)
qc.cx(2, 3)

# Step 3: Measure
qc.measure([0,1,2,3], [0,1,2,3])

# Run
backend = Aer.get_backend('qasm_simulator')
compiled = transpile(qc, backend)
job = backend.run(compiled, shots=1024)

result = job.result()
counts = result.get_counts()

print(counts)
```
