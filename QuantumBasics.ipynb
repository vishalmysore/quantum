1пёҸвғЈ Define a quantum circuit (1 qubit, 1 classical bit)

```python
from qiskit import QuantumCircuit

# 1 qubit, 1 classical bit
qc = QuantumCircuit(1, 1)

print(qc)
```

рҹ”Қ What this actually means

This is Qiskit showing a text diagram of your circuit.

Breakdown:
q: вҶ’ quantum register (your qubits)
c: вҶ’ classical register (your classical bits)
1/ вҶ’ means 1 classical bit exists

рҹ‘ү BUT:

You havenвҖҷt added any gates yet
So nothing is drawn on the wires

2пёҸвғЈ Define a circuit with ONLY 1 qubit (no classical bit)

```python
from qiskit import QuantumCircuit

# Only 1 qubit, no classical bits
qc = QuantumCircuit(1)

print(qc)
```

3пёҸвғЈ Using CNOT (Controlled NOT gate)

```python
from qiskit import QuantumCircuit

# 2 qubits, 2 classical bits
qc = QuantumCircuit(2, 2)

# Put first qubit in |1вҹ©
qc.x(0)

# Apply CNOT: control=0, target=1
qc.cx(0, 1)

print(qc)
```

4пёҸвғЈ Measuring qubits

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1, 1)

# Flip qubit to |1вҹ©
qc.x(0)

# Measure qubit вҶ’ classical bit
qc.measure(0, 0)

print(qc)
```

5пёҸвғЈ Hadamard gate (superposition)

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1, 1)

# Create superposition
qc.h(0)

# Measure
qc.measure(0, 0)

print(qc)
```

Finally  , Run the Simulator

```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit import transpile

# create circuit
qc = QuantumCircuit(1, 1)
qc.h(0)
qc.measure(0, 0)

# simulator
simulator = Aer.get_backend('aer_simulator')

# transpile + run
compiled = transpile(qc, simulator)
job = simulator.run(compiled, shots=1000)

result = job.result()
counts = result.get_counts()

print(counts)
```

This program flips a qubit to 1 , measures it, and runs a simulator to confirm it always outputs 1.

```python
from qiskit_aer import Aer
from qiskit import transpile
qc = QuantumCircuit(1, 1)

# Flip qubit to |1вҹ©
qc.x(0)
qc.measure(0, 0)
sim = Aer.get_backend('aer_simulator')
compiled = transpile(qc, sim)
job = sim.run(compiled, shots=1000)

result = job.result()
print(result.get_counts())
```
