```python
!pip install qiskit
!pip install qiskit-aer
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

# Create circuit
qc = QuantumCircuit(1, 1)
qc.h(0)              # superposition
qc.measure(0, 0)     # measure

# Create simulator
simulator = AerSimulator()

# Compile (transpile) circuit for simulator
compiled_circuit = transpile(qc, simulator)

# Run the circuit
job = simulator.run(compiled_circuit, shots=1000)

# Get results
result = job.result()

# Get counts (how many times 0 or 1 occurred)
counts = result.get_counts()

print(counts) # measure qubit 0 and store the result in classical bit 0
```

```python
!pip install qiskit
!pip install qiskit-aer
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
import numpy as np

# Create a Quantum Circuit with one qubit and one classical bit
# The qubit is named 'q' and the classical bit is named 'c'
qc = QuantumCircuit(1, 1)

# A newly initialized qubit in Qiskit starts in the |0> state by default.
# To represent a specific amplitude for a state, we can apply gates.
# For example, let's put the qubit into an equal superposition state:
# |ψ⟩ = (1/√2)|0⟩ + (1/√2)|1⟩
# This means the probability amplitude for |0⟩ is 1/√2 and for |1⟩ is 1/√2.
# A Hadamard gate (H-gate) achieves this.
qc.h(0) # Apply Hadamard gate to qubit 0

# Explicitly save the statevector at this point in the circuit
qc.save_statevector()

# Add a measurement to store the result in the classical bit
qc.measure(0, 0) # Measure qubit 0 and store the result in classical bit 0

# To 'define' the qubit in this superposition, we just need the circuit.
# We can also visualize the circuit:
print("Quantum Circuit representing our qubit in superposition (with measurement):")
print(qc)

# You can't directly 'print' a qubit's amplitude in Qiskit without simulating it.
# To see the amplitudes, we need to simulate the circuit and get the statevector.
# We explicitly set the method to 'statevector' for AerSimulator.
simulator = AerSimulator(method='statevector')
# Removed the 'transpile' step as it's not strictly necessary for this simulation
job = simulator.run(qc) # Run the original circuit directly
result = job.result()
statevector = result.get_statevector()

print(f"\nStatevector (amplitudes) of the qubit: {statevector}")

# The statevector shows the complex amplitudes for the |0> and |1> states.
# For a single qubit, statevector[0] is the amplitude for |0> and statevector[1] for |1>.
amplitude_0 = statevector[0]
amplitude_1 = statevector[1]

print(f"Amplitude for state |0>: {amplitude_0:.4f}")
print(f"Amplitude for state |1>: {amplitude_1:.4f}")

# Let's verify the probabilities by squaring the magnitudes of these amplitudes
prob_0 = np.abs(amplitude_0)**2
prob_1 = np.abs(amplitude_1)**2

print(f"Probability of measuring |0>: {prob_0:.4f}")
print(f"Probability of measuring |1>: {prob_1:.4f}")
print(f"Sum of probabilities: {prob_0 + prob_1:.4f}")
```

```python
import numpy as np

# Define a complex amplitude for a qubit state
# For example, let's say the amplitude for measuring '0' is (0.6 + 0.8j)
# The sum of squared magnitudes for all possible states must equal 1.
# In this simplified example, we'll just consider one state's amplitude.
amplitude = 0.6 + 0.8j

print(f"The amplitude for a certain state is: {amplitude}")

# Calculate the squared magnitude of the amplitude
# In NumPy, np.abs() computes the magnitude, and we then square it.
probability = np.abs(amplitude)**2

print(f"The squared magnitude (probability) is: {probability:.4f}")

# Verify that the probability is between 0 and 1
if 0 <= probability <= 1:
    print("This probability is valid as it is between 0 and 1.")
else:
    print("This probability is outside the valid range [0, 1].")

# Example with another amplitude
amplitude_2 = 1/np.sqrt(2) + 1/np.sqrt(2) * 1j # Represents a superposition state
probability_2 = np.abs(amplitude_2)**2

print(f"\nAnother amplitude: {amplitude_2:.4f}")
print(f"Its squared magnitude (probability) is: {probability_2:.4f}")
```
